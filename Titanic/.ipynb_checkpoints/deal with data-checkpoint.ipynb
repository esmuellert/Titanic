```python
import numpy as np
import tensorflow as tf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# deal with age lable
train_raw = pd.read_csv("train.csv")
for i in range(891):
    if np.isnan(train_raw.iloc[i,5]):
        if "Miss" in train_raw.iloc[i,3]:
            train_raw.iloc[i,5] = 10
        if "Master" in train_raw.iloc[i,3]:
            train_raw.iloc[i,5] = 4
        else:
            train_raw.iloc[i,5] = 30
```

```python
# save to csv
train_raw.to_csv('train_age.csv')

train_age = pd.read_csv('train_age.csv')
tmp = pd.read_csv("train_1.csv")

tmp['Age'] = train_age.iloc[:,6]
tmp = tmp.drop(columns=['SibSp','Parch'])
tmp.to_csv('train_3.csv')
```

```python
# deal with test sets

test_raw = pd.read_csv("test.csv")
for i in range(418):
    if np.isnan(test_raw.iloc[i,4]):
        if "Miss" in test_raw.iloc[i,2]:
            test_raw.iloc[i,4] = 10
        if "Master" in test_raw.iloc[i,2]:
            test_raw.iloc[i,4] = 4
        else:
            test_raw.iloc[i,4] = 30

# save to csv

tmp = pd.read_csv("test_1.csv")
tmp['Age'] = test_raw.iloc[:,4]
tmp = tmp.drop(columns=['SibSp','Parch'])
tmp.to_csv("test_3.csv")
```
